```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/student_data.csv')
df.info()
```

```python
df = df.drop(['Unnamed: 0'],axis =1)
df
```

```python
df['EthnicGroup'] =df['EthnicGroup'].fillna('No Group')
df.info()
```

```python
df['ParentEduc'].unique()
```

```python
df['ParentEduc']=df['ParentEduc'].fillna('No Education')
df['ParentEduc'].unique()
```

```python
df['TestPrep'].unique()
```

```python
d={'none':'not started','completed':'completed'}
df['TestPrep']=df['TestPrep'].map(d)
df['TestPrep'].unique()
```

```python
df['TestPrep'] = df['TestPrep'].fillna('status pending')
df['TestPrep'].unique()
```

```python
df['ParentMaritalStatus'].unique()
```

```python
df['ParentMaritalStatus']=df['ParentMaritalStatus'].fillna('not updated')
df['ParentMaritalStatus'].unique()
```

```python
df['PracticeSport'].unique()
```

```python
df['PracticeSport']=df['PracticeSport'].fillna('not updated')
df['PracticeSport'].unique()
```

```python
df['IsFirstChild'].unique()
```

```python
d={'yes':'yes','no':'No'}
df['IsFirstChild']=df['IsFirstChild'].map(d)
df['IsFirstChild'].unique()
```

```python
df['IsFirstChild']=df['IsFirstChild'].fillna('not updated')
df['IsFirstChild'].unique()
```

```python
df['NrSiblings'].unique()
```

```python
df['NrSiblings']=df['NrSiblings'].fillna('not updated')
df['NrSiblings'].unique()
```

```python
df['TransportMeans'].unique()
```

```python
df['TransportMeans']=df['TransportMeans'].fillna('not updated')
df['TransportMeans'].unique()
```

```python
df['WklyStudyHours'].unique()
```

```python
df['WklyStudyHours']=df['WklyStudyHours'].fillna('not updated')
df['WklyStudyHours'].unique()
```
